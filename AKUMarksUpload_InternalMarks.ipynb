```python
#print(list_of_rows)

def findReg(regNo):
    #print(regNo)
    R = []
    for one in list_of_rows:
        #print(one[2])
        if one[2].strip() == regNo:
            R = one
            break
    return R

#Enable/Uncomment this block For Internal Theory/Lab and comment out next block

#ctl00_mastercontentplaceholder_gvInternalmark
for i in range(0, row_count-1):
    l = driver.find_elements_by_xpath("//table[@id='ctl00_mastercontentplaceholder_gvInternalmark']/tbody/tr["+str((i+2))+"]/td[2]")
    #print(l[0].text)
    oneData = findReg(l[0].text)
    if len(oneData) != 0:
        #print(oneData)
        x = "{:0>2d}".format(i+2)
        ind = 4                           #Start Index of Marks
        
        #Next three lines are for Attendance Marks
        attndM = driver.find_element_by_id("ctl00_mastercontentplaceholder_gvInternalmark_ctl"+x+"_txtAttendancemark")
        attndM.clear()
        attndM.send_keys(oneData[ind])
        
        #Next three lines are for Class Test Marks
        classM = driver.find_element_by_id("ctl00_mastercontentplaceholder_gvInternalmark_ctl"+x+"_txtClassTestmark")
        classM.clear()
        classM.send_keys(oneData[ind+1])
        
        #Next three lines are for Internal Marks
        eseM = driver.find_element_by_id("ctl00_mastercontentplaceholder_gvInternalmark_ctl"+x+"_txtESEmark")
        eseM.clear()
        eseM.send_keys(oneData[ind+2])
       
        #Next three lines are for Converting Total to Word
        eseM = driver.find_element_by_id("ctl00_mastercontentplaceholder_gvInternalmark_ctl"+x+"_txtMarkWord")
        eseM.clear()
        ESE_Marks=oneData[ind+2]
        if ESE_Marks=="AB":
            ESE_Marks=0
        eseM.send_keys(marksInWords[int(oneData[ind])+int(oneData[ind+1])+int(ESE_Marks)])

        #Displaying entered marks details if successful
        print(oneData," ", marksInWords[int(oneData[ind])+int(oneData[ind+1])+int(ESE_Marks)])
    else:
       print("Error in Input File: "+ l[0].text + " Details Not Found, Check your input file!!")


#End
```
